# 检测器

```python
checkpointPath = './PSENet/checkpoints/checkpoint.pth.tar'
TextDetect = PSEDetect(checkpointPath)

yolo_cfg = 'yolov3-com/cfg/yolov3.cfg'
weightsPath = 'yolov3-com/weights/yolov3.weights'
com_detect = Yolov3Detect(yolo_cfg,weightsPath)
```

```python
#图像
import requests as req
from PIL import Image
from io import BytesIO
import time 

imgDir = '/home/datalab/ex_disk/work/shengdan/jupyter/pytorch/OCR-dectect/seglink/material/'
imagenames = os.listdir(imgDir)
imgpaths = [os.path.join(imgDir,i) for i in imagenames]

outDir = '/home/datalab/ex_disk/work/shengdan/dataset/posterout'
for i,imgpath in enumerate(imgpaths):
    
    #检测文本
    img_cv2 = cv2.imread(imgpath)
    
    t0 = time.time()
    img,org_img = create_input(img_cv2)
    bboxes = predict(img,org_img,TextDetect.model)
    TDT = time.time()-t0
    print('Text detect using time:',TDT)
    
    #画text
    org_img = org_img.numpy().astype('uint8')[0]
    text_box = org_img.copy()
    for bbox in bboxes:
        cv2.drawContours(text_box, [bbox.reshape(4, 2)], -1, (0, 255, 0), 2)

    #检测主体
    t1 = time.time() 
    yolov3input = createInput (img_cv2)
    res = detect(com_detect.model,yolov3input)
    CDT = time.time()-t0
    print('Text detect using time:',CDT)
    
    print('ALL USING TIME:',TDT+CDT)
    if res is not None:
        for x1, y1, x2, y2, conf, cls_conf, cls in res:
            # Add bbox to the image
            plot_one_box([x1, y1, x2, y2], text_box, label='', color=(100,100,30))

    savepath = os.path.join(outDir,os.path.basename(imgpath))
    cv2.imwrite(savepath,text_box)

```
